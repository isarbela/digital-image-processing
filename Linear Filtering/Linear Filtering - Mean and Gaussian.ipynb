## Imports

```python
import skimage
import skimage.io
import numpy as np
from matplotlib import pyplot as plt
from scipy.ndimage import convolve
```

## Mean Filter

```python
def mean_filter(img, kernel_size):
    """
    apply a mean filter to an image
    
    img - input image
    kernel_size = kernel size 
    
    """
    
    # Padding image to handle borders
    padded_img = np.pad(img, kernel_size//2, mode='reflect')
    
    filtered_img = np.zeros_like(img)
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            window = padded_img[i:i + kernel_size, j:j +kernel_size]
            mean = np.mean(window)
            filtered_img[i, j] = int(mean)
    
    return filtered_img
```

## Gaussian Filter

```python
def gaussian_kernel(l, sig):
    """
    creates a 2D gaussian kernel
    
    l - kernel size (lxl)
    sig - sigma
    
    """
    ax = np.linspace(-(l - 1) / 2., (l - 1) / 2., l)
    kernel_1d = np.exp(-0.5 * np.square(ax) / np.square(sig))
    kernel_2d = np.outer(kernel_1d, kernel_1d)
    return kernel_2d/kernel_2d.sum()


def gaussian_filter(img, kernel_size, sig):

    gauss_filter = gaussian_kernel(kernel_size, sig)
    print(gauss_filter)
    filtered_img = convolve(img, gauss_filter)
    
    return filtered_img
        
```

```python
def main():
    image = skimage.io.imread('lena.png')
    image = image.astype(np.uint8)
    
    plt.title('Original Image',  
                                     fontweight ="bold") 
    plt.imshow(image, 'gray')
    plt.figure()
    
    # Apply gaussian noise to the image
    sigma = 10
    noise = np.random.normal(0, sigma, (image.shape[0], image.shape[1]))
    img_noise = image + noise
    img_noise[np.where(img_noise > 255)] = 255
    img_noise[np.where(img_noise < 0)] = 0
    
    
    plt.title('Noise Image',  
                                     fontweight ="bold") 
    plt.imshow(img_noise, 'gray')
    plt.figure()
    
    # Filtering image using a 3x3 mean filter
    filtered_img_mean = mean_filter(img_noise, 3)
    
    plt.title('After mean filter noise image',  
                                     fontweight ="bold") 
    plt.imshow(filtered_img_mean, 'gray')
    plt.figure()
    
    
     # Filtering image using a 5x5 sigma 1. gaussian filter
    filtered_img_gauss = gaussian_filter(img_noise, 5, 1.)
    
    plt.title('After gaussian filter noise image',  
                                     fontweight ="bold") 
    plt.imshow(filtered_img_gauss, 'gray')
    plt.figure()
    
if __name__ == "__main__":
    main()    
```
